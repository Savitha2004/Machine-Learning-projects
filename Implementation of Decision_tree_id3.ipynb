```python
redata=data
party={'republican':0,'democrat':1}
vote={'y':1,'n':0,'?':0}
for col in data.columns:
  if col !='party':
    data[col]=data[col].map(vote)
data['party']=data['party'].map(party)
data
```

```python
redata.head()
```

```python
X=data.iloc[:, 1:17].values
y=data['party']=data.iloc[:,0].values
from sklearn.model_selection import KFold
kf=KFold(n_splits=5)
for train_index,test_index in kf.split(X,y):
  X_train,X_test=X[train_index],X[test_index]
  y_train,y_test=y[train_index],y[test_index]
```

```python
X_train.shape,X_test.shape,data.shape,y_train.shape,y_test.shape
```

```python
from classic_ID3_decision_tree import DecisionTreeClassifier
id3=DecisionTreeClassifier()
id3.add_features(data,'party')
print(id3.features)
id3.gini_index(X_train,y_train)
y_pred=id3.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,y_pred)
formatted_accuracy="{:.2f}".format(100*accuracy)
print(f'Accuracy:',formatted_accuracy)
```

```python
from sklearn.tree import plot_tree
from sklearn.tree import DecisionTreeClassifier
plt.figure(figsize=(10,3),dpi=500)
model=DecisionTreeClassifier(max_depth=3)
model.fit(X_train,y_train)
plot_tree(model,feature_names=data.columns,filled=True)
```
